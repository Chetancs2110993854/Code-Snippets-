```python
##Group by on normal dataframe
import pandas as pd

# Sample DataFrame
data = {
    'Company': ['AAPL', 'AAPL', 'MSFT', 'MSFT', 'GOOG', 'GOOG'],
    'Date': ['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02'],
    'Adj Close': [150, 152, 230, 232, 2750, 2780],
    'Volume': [100000, 120000, 130000, 140000, 20000, 21000]
}

df2 = pd.DataFrame(data)
df2
```

```python
df2.groupby('Company').sum()
df2.groupby(['Company','Date']).mean()
df2.groupby('Company').count()
df2.groupby('Company').describe()
```

```python
#transform
df2.groupby(['Company','Date'])[['Volume','Adj Close']].transform('sum')
```

```python
df2
```

```python
df2.groupby('Company')['Adj Close'].apply(lambda x : x.max()- x.min())
```

```python
df2.groupby('Company')['Adj Close'].transform(lambda x : x.max()- x.min())
```

```python
df2.groupby('Company')['Volume'].apply(lambda x : x / 100)
```

```python
df2.groupby(['Company', 'Date'])['Volume'].apply(lambda x : x / 100)
```

```python
df2.groupby(['Company', 'Date'])['Volume'].transform(lambda x : x / 100)# use transform if you want to make a new coloumn
```
